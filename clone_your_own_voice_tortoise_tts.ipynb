# **Creating Folders**

```python
import os

def check_and_create_folder(folder_path):
    if not os.path.exists(folder_path):
        try:
            os.makedirs(folder_path)
            print(f"Folder '{folder_path}' did not exist and has been created.")
        except OSError as e:
            print(f"Error creating folder '{folder_path}': {e}")
    else:
        print(f"Folder '{folder_path}' already exists.")

folder_path = ["/content/Training_Audios", "/content/Txt_Files"]  # Replace with the desired folder path
for folder in folder_path:
    check_and_create_folder(folder)
```

Downloading Training_Audio Files Mp3





```python
import requests

def download_mp3(url, save_path):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check for any errors in the HTTP request

        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {url} to {save_path}")

    except requests.exceptions.RequestException as e:
        print(f"Error downloading {url}: {e}")

audio_list = ["https://raw.githubusercontent.com/PopoolaIbrahim/PDF-to-Video/main/ElevenLabs_Matthew_1.mp3","https://raw.githubusercontent.com/PopoolaIbrahim/PDF-to-Video/main/ElevenLabs_Matthew_2.mp3"]
path_list = ["/content/Training_Audios/ElevenLabs_Matthew_1.mp3","/content/Training_Audios/ElevenLabs_Matthew_2.mp3"]
for mp3_url, save_path in zip(audio_list, path_list) :
  download_mp3(mp3_url, save_path)
```

Converting Mp3 to Wav

```python
!pip install pydub
from pydub import AudioSegment
import os

def mp3_to_wav(mp3_file, wav_file):
    # Load the MP3 file
    audio = AudioSegment.from_mp3(mp3_file)

    # Export the audio as WAV
    audio.export(wav_file, format="wav")

for audio in os.listdir("/content/Training_Audios"):
  mp3_file = f"/content/Training_Audios/{audio}"  # Replace with the path to your MP3 file
  wav_file = f"/content/Training_Audios/{audio[:-3]}" + "wav"
  mp3_to_wav(mp3_file, wav_file)
  print(f"Conversion from {mp3_file} to {wav_file} complete.")
```
